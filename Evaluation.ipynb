```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt

# Load the csv file
df1 = pd.read_csv('C:\\Users\\user\\Desktop\\KEJA PALI\\fyp\\fyp\\downloaded csv\\FoodData_Central\\food.csv')
df2 = pd.read_csv('C:\\Users\\user\\Desktop\\KEJA PALI\\fyp\\fyp\\downloaded csv\\FoodData_Central\\branded_food.csv')
df3 = pd.read_csv('C:\\Users\\user\\Desktop\\KEJA PALI\\fyp\\fyp\\downloaded csv\\FoodData_Central\\nutrition.csv')
```

```python
def categorical_eval(df1, feature, add_text):
    freqs = df1[feature].value_counts()
    print(freqs)
    plt.figure(figsize=(10,6))
    sns.barplot(x=freqs.index, y=freqs.values)
    plt.title(feature + add_text)
    plt.grid()
    plt.xticks(rotation=90)
    plt.show()
```

```python
categorical_eval(df1, 'data_type', ' - Frequencies')
```
